# IMDB Data

### One hot enconding

##### Import library and dataset

```python
import pandas as pd
import matplotlib as plt

pd.set_option('display.max_columns', None)

data = pd.read_csv("E:\Kuliah\Proyek Data Science\ProyekDS-Tugas1\IMDB-Movie-Data.csv")
data.head()
```

##### Column: Genre

```python
#empty set to store all genre
allGenre = set()

#iterate through the 'Genre' column and fill the set
for i in data.Genre:
    genreArray = i.split(',')
    for j in genreArray:
        allGenre.add(j)
```

```python
#create array of zeros with dataframe number of rows as length
zeros = [0] * len(data.index)
#this array is used to create a new column for each Genre in the allGenre set

#create new column in the dataset
for n in allGenre:
    data['{}'.format(n)] = zeros
    
data.head()
```

```python
#assign 0 or 1 to the new columns according to the Genre of each film
j = 0
while j < len(data.Genre):
    genre = data.Genre[j]
    for n in allGenre:
        if genre.find(n) != -1:
            data['{}'.format(n)][j] = 1
        else:
            data['{}'.format(n)][j] = 0
    j += 1
```

##### K-Modes clustering

```python
#select the genre column as features
array_features = list(allGenre)
data_features = data[array_features]

#convert sliced dataframe to array
import numpy as np
array_data_features = np.array(data_features.values)
```

```python
#import library
from sklearn.cluster import KMeans

#run the algorithm 
n = 2
while n < 21:
    kmodes_model = KMeans(n_clusters = n).fit(array_data_features)

    cluster_object = kmodes_model.labels_

    centroids = kmodes_model.cluster_centers_

    from sklearn.metrics import silhouette_score
    coef_score = silhouette_score(array_data_features, cluster_object)
    
    #print("centroids:\n", centroids)
    print("{} - Koefisien Silhouette:".format(n), coef_score, end="\n\n")
    n += 1

```

```python
from sklearn.cluster import AgglomerativeClustering
n = 2
while(n < 11):
    #Buat model Agglomerative
    agglo_model = AgglomerativeClustering(distance_threshold=None, n_clusters=n).fit(array_data_features)
    
    #Array untuk label setiap baris
    klaster_objek = agglo_model.labels_
    
    #Menghitung nilai Koefisien Silhouette
    coef_score = silhouette_score(array_data_features, klaster_objek)
    print("{} - Koefisien Silhouette:".format(n), coef_score, end="\n\n")
    n += 1
```
